The graph below:
* uses ax.text to plot labels on the series
    - the first argument is the percentage of the figure width, which indicates the left location of the text
    - the second argument is the percentage of the figure height, which indicates the bottom of the text
    - the parameter setting, transform=ax.transAxes, permits the text lcoation coordinates to be specified as stated in the above two points 
    - the x and y coordinates need to be determined pretty much on a a trial and error basis
    - I've used the ax.set_xlim(0,18) statement to leave some room on the right side of the graph for the labels
* uses ax.axvspan(beg_index, end_index) to highlight the second year of the data
    - the beg_index is the beginning index in the DataFrame where the highlight begins
    - the end_index is the ending index of the DataFrame where the highlight ends
    - Thus, this statement, ax.axvspan(4,7), highlights the second year of data in the DataFrame, that is, the 4th through 7th quarters
    - I had shown the patches/rectangle approach in class, but Bailey Mitchell, found this approach online.  It is much better.

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(df1.index,df1.Product1Demand, color = 'green')
ax.plot(df2.index,df2.Product2Demand, color = 'blue')
ax.set_xlim(0,18)                                       # this leaves some room on the right for the text labels
ax.axvspan(4,7,color = 'grey', alpha = 0.2)
fig.set_size_inches(12,8)
ax.text(0.85,0.75,'Product 1', fontsize = 16, color = 'green', transform=ax.transAxes)
ax.text(0.85,0.87,'Product 2', fontsize = 16, color = 'blue', transform=ax.transAxes)
```
